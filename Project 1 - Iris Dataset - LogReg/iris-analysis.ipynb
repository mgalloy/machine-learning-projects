```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
dataset = datasets.load_iris()
```

```python
from sklearn.linear_model.LinearRegression
```

```python
print dataset.DESCR
```

```python
print dataset.target_names[dataset.target]
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=10)
#model = sklearn.linear_model.SGDClassifier(loss='log')
model.fit(dataset.data, dataset.target)
print(model)
```

```python
expected = dataset.target
predicted = model.predict(dataset.data)
print predicted
```

```python
predicted_proba = model.predict_proba(dataset.data)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15, 3.0))
cmap = cm.Set1
for i in range(3):
    axes.plot(predicted_proba[:, i], c=cmap(i / 2.0))
```

```python
print(metrics.classification_report(expected, predicted))
```

```python
print(metrics.confusion_matrix(expected, predicted))
```
